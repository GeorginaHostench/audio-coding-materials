```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft
from scipy.signal import get_window
```

```python
# Sine window
M = 64
n = np.arange(M)
w0 = np.sin((n+0.5)*np.pi/(M))
w = w0 / ( np.sum(w0))
plt.plot(w0)
plt.show()
X = fft(w, 1024)
mX=20*np.log10(abs(X[:X.size//2]))
plt.plot(mX)
plt.axis([0,mX.size,-90,0])
plt.show()
```

```python
# Hanning window
M = 64
n = np.arange(M)
w0 = get_window('han',M)
w = w0 / ( np.sum(w0))
plt.plot(w0)
plt.show()
X = fft(w, 1024)
mX=20*np.log10(abs(X[:X.size//2]))
plt.plot(mX)
plt.axis([0,mX.size,-90,0])
plt.show()
```

```python
# Rectanguar window
M = 64
n = np.arange(M)
w0 = get_window('boxcar',M)
w = w0 / ( np.sum(w0))
plt.plot(w0)
plt.show()
X = fft(w, 1024)
mX=20*np.log10(abs(X[:X.size//2]))
plt.plot(mX)
plt.axis([0,mX.size,-50,0])
plt.show()
```
